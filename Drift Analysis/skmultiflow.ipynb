## Churn data

```python
# Imports
from skmultiflow.drift_detection import KSWIN

ds = churn_df["CreditScore"]
window_size = 100

kswin = KSWIN(alpha=0.05, window_size=window_size, data=ds[:8000])

detections = []
for i in range(8000,10000):
        kswin.add_element(ds[i])
        if(kswin.window.shape[0] == window_size):
            print("Drift velocity at " + str(i) + " : " + str( (1 - kswin.p_value) * 100) + " %" )        
        if kswin.detected_change():
            print('Change has been detected in data: ' + str(ds[i]) + ' - of index: ' + str(i))
            print("\tDrift velocity at " + str(i) + " : " + str( (1 - kswin.p_value) * 100) + " %" )   
            kswin.reset()
            kswin.p_value=1
            detections.append(i)     
        
print("Number of detections: "+str(len(detections)))
```

```python
from sklearn.model_selection import train_test_split
churn_train, churn_eval = train_test_split(ds, test_size=0.2)

churn_train = churn_train.reset_index(drop=True)
churn_eval = churn_eval.reset_index(drop=True)
```

```python
#abrupt shift 

churn_eval[0:500] = 0
churn_eval[500:1000] = 500
churn_eval[1000:1500] = 1000
churn_eval[1500:2000] = 2000

kswin = KSWIN(alpha=0.005, window_size=100) #, data=churn_train)
for i in range(len(churn_train)):
        kswin.add_element(churn_train[i])
        
detections = []
for i in range(len(churn_eval)):
        kswin.add_element(churn_eval[i])
        print("Drift velocity at " + str(i) + " : " + str( (1 - kswin.p_value) * 100) + " %"  + " - p-value:" + str(kswin.p_value))
        if kswin.detected_change():
            print('Change has been detected at Index: ' + str(i) + " - Data : " + str(churn_eval[i]))
            kswin.reset()
            #print("\r KSWINReject Null Hyptheses")
            detections.append(i)
            
print("Number of detections: "+str(len(detections)))
```

## Gradual shift

```python
max_credit_score = max(churn_train)

new_max_credit_score1 = int( max_credit_score + ( 10 * 100 / max_credit_score ))
new_max_credit_score2 = int( max_credit_score + ( 20 * 100 / max_credit_score ))
new_max_credit_score3 = int( max_credit_score + ( 30 * 100 / max_credit_score ))
new_max_credit_score4 = int( max_credit_score + ( 40 * 100 / max_credit_score ))

churn_eval[0:500] = np.random.randint(new_max_credit_score / 2,high=new_max_credit_score1, size=500)
churn_eval[500:1000] = np.random.randint(new_max_credit_score / 2,high=new_max_credit_score2, size=500)
churn_eval[1000:1500] = np.random.randint(new_max_credit_score / 2,high=new_max_credit_score3, size=500)
churn_eval[1500:2000] = np.random.randint(new_max_credit_score / 2,high=new_max_credit_score4, size=500)
```

```python
kswin = KSWIN(alpha=0.005, window_size=100) #, data=churn_train)
for i in range(len(churn_train)):
        kswin.add_element(churn_train[i])
        
detections = []
for i in range(len(churn_eval)):
        kswin.add_element(churn_eval[i])
        print("Drift velocity at " + str(i) + " : " + str( (1 - kswin.p_value) * 100) + " %" )
        if kswin.detected_change():
            print('Change has been detected at Index: ' + str(i) + " - Data : " + str(churn_eval[i]))
            kswin.reset()
            #print("\r KSWINReject Null Hyptheses")
            detections.append(i)
            
print("Number of detections: "+str(len(detections)))
```
