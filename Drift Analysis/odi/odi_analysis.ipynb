```python
from espncricinfo.summary import Summary
from espncricinfo.match import Match

import numpy as np 
import pandas as pd
```

```python
first_match_id = 64148
last_match_id = 1223957
```

```python
odi_data = []

start =  1150646 #1140512 #1129846 #last_match_id - 100000 #99000 #75320 #67523 #66808 #66388
last = start + 10000
for i in np.arange(start, last+1):
    try :
        m = Match(i)
    except Exception as e:
        print(i, "E")
        continue
    print(i)
    if m.match_json()["international_class_id"] == "2" and m.match_json()["result_short_name"]!= "aban" and m.match_json()["result_short_name"]!= "canc" and m.match_json()["winner_team_id"] != "0" :
        print(m.match_json()['date'])
        match = m.match_json()
        if len(m.innings) == 0:
            continue
        innings1, innings2 = m.innings
        odi_data.append([
            match['date'], match["team1_name"], match["team2_name"], 
            match["winner_team_id"], match["team1_id"], match["team2_id"], 
            match["cancelled_match"], match["floodlit"],
            innings1["ball_limit"], innings2["ball_limit"],
            innings1["over_limit"], innings2["over_limit"],            
            innings1['runs'], innings2['runs'],
            innings1["wickets"], innings2["wickets"],
            innings1["run_rate"], innings2["run_rate"]
        ])
```

```python
df = pd.DataFrame(odi_data, columns=["Date", "Team1", "Team2", "Team1_id", "Team2_id", "Winning_team_id",
        "cancelled_match", "floodlit", "innings1_ball_limit", "innings2_ball_limit", 
        "innings1_over_limit", "innings2_over_limit", "innings1_runs", "innings2_runs",
        "innings1_wickets", "innings2_wickets", "innings1_run_rate", "innings2_run_rate"
])
df.to_csv("odi_data6.csv")
```
